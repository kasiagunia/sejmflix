```python
for file_path in os.listdir(f'./data'):
    if 'GLO' not in file_path:
        continue
    df = pd.read_csv(f'./data/{file_path}')
    
    first_vote_id = df['vote_id'].min()
    last_vote_id = df['vote_id'].max()
    
    df1 = df[df['vote_id'] == first_vote_id]
    df2 = df[df['vote_id'] == last_vote_id]
    m1 = df1['Nrleg'].unique().size
    m2 = df2['Nrleg'].unique().size
    
    g1 = df1.groupby('Klub', as_index=False)['Nrleg'].nunique()
    g2 = df2.groupby('Klub', as_index=False)['Nrleg'].nunique()
    
    df_res = g1.merge(g2, how='outer', on='Klub', suffixes=('_start', '_end'))
    df_res['Nrleg_start_share'] = df_res['Nrleg_start'] / m1 * 100
    df_res['Nrleg_end_share'] = df_res['Nrleg_end'] / m2 * 100
    print(file_path)
    display(df_res)
```

```python
def start_end_share(df):
    first_vote_id = df['vote_id'].min()
    last_vote_id = df['vote_id'].max()
    
    df1 = df[df['vote_id'] == first_vote_id]
    df2 = df[df['vote_id'] == last_vote_id]
    m1 = df1['Nrleg'].unique().size
    m2 = df2['Nrleg'].unique().size
    
    g1 = df1.groupby('Klub', as_index=False)['Nrleg'].nunique()
    g2 = df2.groupby('Klub', as_index=False)['Nrleg'].nunique()
    
    df_res = g1.merge(g2, how='outer', on='Klub', suffixes=('_start', '_end'))
    df_res['Nrleg_start_share'] = df_res['Nrleg_start'] / m1 * 100
    df_res['Nrleg_end_share'] = df_res['Nrleg_end'] / m2 * 100
    return df_res
```
